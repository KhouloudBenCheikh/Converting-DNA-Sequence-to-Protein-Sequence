**From DNA to Protein Process**

**1) Transcription (DNA --> RNA):** From the coding Strand replace letter T with U

image.png

```python
#Install BioPython Library
pip install biopython
```

```python
#import Library
from Bio.Seq import Seq
from Bio.SeqUtils import GC
```

```python
dna_coding_strand = Seq("ATGGCCATTGTAATGGGCCGCTGAAAGGGTGCCGATAG")
```

```python
#Complement DNA
dna_template_strand = dna_coding_strand.reverse_complement()

print(dna_coding_strand)
print(dna_template_strand)
```

```python
#Transcription
messenger_rna = dna_template_strand.transcribe()
print('DNA: ', dna_template_strand)
print('RNA: ',messenger_rna)
```

*** Back Transcription: (RNA --> DNA)**

```python
dna_template_strandd = messenger_rna.back_transcribe()
print('RNA: ',messenger_rna)
print('DNA: ', dna_template_strandd)
```
